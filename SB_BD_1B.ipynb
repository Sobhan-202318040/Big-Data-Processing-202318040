```python
import requests
import csv
import pandas as pd

url = "https://real-time-amazon-data.p.rapidapi.com/search"

querystring = {"query": "Medicine", "page": "1", "country": "US", "category_id": "aps"}

headers = {
    "X-RapidAPI-Key": "5fde53e60dmshee72ae95caecfb4p1ab381jsn408b9da1ff62",
    "X-RapidAPI-Host": "real-time-amazon-data.p.rapidapi.com",
}

response = requests.get(url, headers=headers, params=querystring)

if response.status_code == 200:
    data = response.json().get("data", {})


    if "products" in data:
        products = data["products"]

        headers = products[0].keys() if products else []
        rows = [product.values() for product in products] if products else []

        csv_file_path = 'amazon_products.csv'
        with open(csv_file_path, 'w', newline='', encoding='utf-8') as csv_file:
            csv_writer = csv.writer(csv_file)

            csv_writer.writerow(headers)

            csv_writer.writerows(rows)

        print(f'Data successfully converted and saved to {csv_file_path}')

    else:
        print("No product information found in the API response.")
else:
    print(f"Failed to retrieve data. Status Code: {response.status_code}")
    print(response.text)
```
