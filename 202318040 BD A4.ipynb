
TASK-1: Generate 100 random numbers in range 0 to 10 using numpy randint function with the seed set to 10. Create a RDD using the paral-
lelize function using data generated in previous step. Calculate the fre-
quency of each number (0 - 10) using appropriate function of RDD.

```python
from pyspark import SparkContext
import numpy as np

sc = SparkContext.getOrCreate()

#Generating 100 random numbers using NumPy
np.random.seed(10)
random_numbers = np.random.randint(0, 11, 100)

#Creating RDD from random generated numbers
rdd = sc.parallelize(random_numbers)

#Finding the frequency of each number using appropriate function of RDD
number_frequency = rdd.map(lambda x: (x, 1)) \
                     .reduceByKey(lambda a, b: a + b) \
                     .sortByKey()

result = number_frequency.collect()
for num, freq in result:
    print(f"Number {num}: Frequency {freq}")

sc.stop()
```

TASK-2: In this task you will calculate the frequency of each word in text8 dataset mentioned above. Create a RDD using the text8 dataset. Use appropriate functions of the RDD to get the word frequencies. Fil ter the RDD using appropriate function to get the frequencies of words containing the letter ’a’.

```python
sc = SparkContext("local", "RDD Task 2")

text_rdd = sc.textFile("/content/text8")

#Splitting text into words and count
word_count_rdd = text_rdd.flatMap(lambda line: line.split()) \
                         .map(lambda word: (word.lower(), 1)) \
                         .reduceByKey(lambda a, b: a + b)

#Filter the RDD to get frequencies of words containing the letter 'a'
word_count_with_a_rdd = word_count_rdd.filter(lambda x: 'a' in x[0])

result = word_count_with_a_rdd.collect()
for word, freq in result:
    print(f"Word: {word}, Frequency: {freq}")

sc.stop()
```

DataFrame Task: Create a Spark dataframe using the iris json data men tioned above. Calculate Pearson Correlation between the columns petal Length and petalWidth using the appropriate dataframe API. Show the columns sepalLength, sepalWidth and species for the rows of data that has petalLength greater than or equal to 1.4 using the appropriate dataframe API.

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, corr

spark = SparkSession.builder \
    .appName("Iris DataFrame Task") \
    .getOrCreate()

iris_df = spark.read.json("/content/iris.json")

#Calculating Pearson correlation between petalLength and petalWidth
correlation = iris_df.corr("petalLength", "petalWidth")

print("Pearson correlation between petalLength and petalWidth:", correlation)

#Filtering the DataFrame based on the condition petalLength >= 1.4
filtered_df = iris_df.filter(col("petalLength") >= 1.4)

#Showing the columns sepalLength, sepalWidth, and species for the filtered rows
filtered_df.select("sepalLength", "sepalWidth", "species").show()

spark.stop()
```
